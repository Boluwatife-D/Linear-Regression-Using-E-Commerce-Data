### BUILDING THE LINEAR REGRESSION ALGORITHM

```python
# Before we build the machine learning algorithm, we have to split the data to training and testing data first

# First we create two variables representing the dependent and independent variables
X = customers.drop(['Yearly Amount Spent', 'Email', 'Address', 'Avatar'], axis = 1)
X
```

```python
Y = customers['Yearly Amount Spent']
Y
# Where X represents the independent variables and 
# Y represent the dependent variable
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size = 0.30, random_state = 101)
```

### TRAINING THE MODEL

```python
# To import LinearRegression from sklearn.linear_model
from sklearn.linear_model import LinearRegression
```

```python
# Creating an instance of a LinearRegression() model named lm.
lm = LinearRegression()
```

```python
lm.fit(X_train, Y_train)
```

```python
# To see the coefficients of the model
coeff_df = pd.DataFrame(lm.coef_, X.columns, columns = ['Coefficient'])
coeff_df
```

```python
#To print out the intercept
print(lm.intercept_)
```

```python
# Our fitted model is then given as 
# Yearly amount spent = -1047.9327822502385 + 25.981550(Avg. Session Length)  + 38.590159(Time on App) + 0.190405(Time on Website) + 61.279097(Length of Membership)
# Now that we have our fitted model, we then go ahead and use the test data on it to see the accuracy of our model
predictions = lm.predict(x_test)
predictions
```

```python
# To see if our model is accurate we create a scatter plot of the predicted independent variables against the original ones
plt.scatter(y_test,predictions)
```

### EVALUATING THE FITTED MODEL

```python
# To import metrics that will be used to measure model accuracy
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

### RESIDUALS

```python
# To get the residuals
residuals = predictions - y_test
residuals
```

```python
# Plotting the residuals should give us a normally distributed data
sns.distplot(residuals)
```
